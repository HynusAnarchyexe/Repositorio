```python
#Cargar las librerias necesrias

import pandas as pd # Manejo y análisis de estructuras de datos
import numpy as np # Cálculo numérico y el análisis de datos
import seaborn as sns # Creación gráficos estadísticos
import matplotlib.pyplot as plt # Creación de gráficos en dos dimensiones

# librerías de para implementar modelos de ciencia de datos
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#from sklearn.linear_model import LinearRegression -> Librería para Regresión Lineal
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# lectura de datos en Python y lo cargamos en la variable Datos

Data = pd.read_csv('./heart_cleveland_upload.csv')
```

```python
# Realizamos lectura de los Datos para el análisis exploratorio

Data.head(50)
```

```python
# Descripción de los datos
Data.describe()
```

```python
#Identificación de Datos Faltantes y Atípicos

plt.figure(figsize=(6, 3))
sns.boxplot(x=Data['oldpeak'])
plt.title('Identificación de Datos Faltantes y Atípicos', fontsize=10 )
```

```python
#Revisión de datos faltantes

print("Cantidad de Registros con cero en Dolor de pecho: ",Data['cp'].isin([0]).sum())
print("Cantidad de Registros con cero en Medición electrocardiográfica en reposo: ",Data['restecg'].isin([0]).sum())
print("Cantidad de Registros con cero en Angina inducida por el ejercicio: ",Data['exang'].isin([0]).sum())
print("Cantidad de Registros con cero en Depresión del ST inducida por el ejercicio en relación con el reposo: ",Data['oldpeak'].isin([0]).sum())
print("Cantidad de Registros con cero en Cantidad de vasos principales: ",Data['ca'].isin([0]).sum())
```

```python
# Reemplazar todos los ceros con el valor medio (promedio) de la columna.

Data['cp']=Data['cp'].replace(0,Data['cp'].mean())
Data['restecg']=Data['restecg'].replace(0,Data['restecg'].mean())
Data['exang']=Data['exang'].replace(0,Data['exang'].mean())
Data['oldpeak']=Data['oldpeak'].replace(0,Data['oldpeak'].mean())
Data['ca']=Data['ca'].replace(0,Data['ca'].mean())
```

```python
# Se revisan nuevamente los datos faltantes

print("Cantidad de Registros con cero en Dolor de pecho: ",Data['cp'].isin([0]).sum())
print("Cantidad de Registros con cero en Medición electrocardiográfica en reposo: ",Data['restecg'].isin([0]).sum())
print("Cantidad de Registros con cero en Angina inducida por el ejercicio: ",Data['exang'].isin([0]).sum())
print("Cantidad de Registros con cero en Depresión del ST inducida por el ejercicio en relación con el reposo: ",Data['oldpeak'].isin([0]).sum())
print("Cantidad de Registros con cero en Cantidad de vasos principales: ",Data['ca'].isin([0]).sum())
```

```python
# Identificar Datos Atípicos

plt.figure(figsize=(6, 3))
sns.boxplot(x=Data['oldpeak'])
plt.title('Identificación de Datos Atípicos', fontsize=10 )
```

```python
#Conteo de Datos Atípicos

nivel_minimo=2
nivel_maximo=3
Total_Atipicos=((Data['oldpeak']<nivel_minimo)|(Data['oldpeak']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
# Filtra Datos Atípicos

Data=Data[(Data['oldpeak']<nivel_maximo)&(Data['oldpeak']>nivel_minimo)]
```

```python
# Validar si se removieron los datos atípicos

Total_Atipicos=((Data['oldpeak']<nivel_minimo)|(Data['oldpeak']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
#Revisión del tipo de datos y datos faltantes

Data.info()
```

```python
# Total de personas con enfermedades cardíacas

Data['condition'].value_counts()
```

```python
Data[['condition']].head(25)
```

```python
# Porcentaje de datos para entrenamiento (train) = 80% , y para prueba (test) = 20%

# Se crean variables X y Y para almacenar la información del entrenamiento

# Se cargan los datos de train sin los datos de la columna Diabetico 
X = Data.drop('condition',axis=1)

# Se cargan los datos de train con los datos de la la columna Diabetico
Y = Data['condition'] 

# La función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state= 0)
```

```python
print (X_train)
```

```python
print (Y_train)
```

```python
# Construir el modelo de regresión logística 

modelo_regresion = LogisticRegression(solver = "liblinear")
modelo_regresion.fit (X_train , Y_train)   

# Crea el modelo de prediccion
ModeloPrediccion = modelo_regresion.predict(X_test)
```

```python
print (Y_test)
```

```python
print (ModeloPrediccion)
```

```python
# Desempeño del modelo de regresión

print(classification_report(Y_test,ModeloPrediccion))
```

```python
# Matriz de Confusión del modelo

cm=confusion_matrix(Y_test,ModeloPrediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Curva de precisión del modelo

from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test, ModeloPrediccion)
plt.show()
```
